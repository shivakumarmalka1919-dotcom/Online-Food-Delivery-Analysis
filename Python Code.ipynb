# Outliers Analysis

```python
import numpy as np
```

```python
numeric_data=df.select_dtypes(include=np.number)
```

```python
for i in numeric_data:
    q1=numeric_data[i].quantile(0.25)
    q2=numeric_data[i].quantile(0.50)
    q3=numeric_data[i].quantile(0.75)
```

```python
IQR=q3-q1
```

```python
Uw=q3+1.5*IQR
Lw=q1-1.5*IQR
```

```python
Outliers=((numeric_data[i]>Uw)|(numeric_data[i]<Lw))
```

```python
print(Outliers.sum())
```
